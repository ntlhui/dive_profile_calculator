```python
from dive_profile_calculator.profile import DiveProfile, DiverConfiguration
import matplotlib.pyplot as plt
```

```python
gue_configuration = DiverConfiguration(13, 232)
gue_ascent = DiveProfile.compute_ssi_ascent(18, gue_configuration)
```

```python
depths = [point.depth for point in gue_ascent]
times = [point.timestamp.total_seconds() / 60 for point in gue_ascent]
```

```python
print(times, depths)
```

```python
plt.plot(times, depths)
plt.grid()
plt.gca().invert_yaxis()
plt.ylabel('Depth (msw)')
plt.xlabel('Time (min)')
```
